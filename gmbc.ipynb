Physical fitness is  the ability of an individual to efficiently carry out daily tasks without experiencing undue fatigue, allowing them to still enjoy their leisure time. Physical fitness can be categorized into three groups: 1) static fitness, which refers to a person's state of being free from illness, 2) dynamic fitness, meaning the ability to work efficiently without requiring specific skills, such as walking, lifting, etc., and 3) motor fitness, which is the ability to perform work with high skills and agility (Kushartanti, Wara). Physical condition plays a crucial role in achieving good performance because it serves as the primary foundation for living a well-functioning life and attaining happiness. A good level of physical fitness is necessary for carrying out daily life, thus requiring a good cardiovascular endurance (VO2Max). The higher one's cardiovascular endurance, the better their physical fitness. Therefore, it is important for each individual to know their cardiovascular endurance at all times to maintain their endurance. This case study aims to facilitate individuals in determining their cardiovascular endurance by only inputting several supporting variables efficiently.isien.

```python
X = dfmodel.drop(['vo2max'], axis = 1)
y = dfmodel['vo2max']
```

Untuk modelling, studi kasus ini menggunakan model Navis Bayes.  "Naive Bayes" merujuk pada asumsi sederhana bahwa setiap pasangan fitur adalah independen, yang pada kenyataannya mungkin tidak selalu terjadi. Namun, meskipun asumsi ini terlalu sederhana, model Naive Bayes sering kali memberikan hasil yang sangat baik dalam praktiknya. 

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X, y)

predicted_navis_baye2 = model.predict(X)
```

```python
y_predict3 = pd.DataFrame(predicted_navis_baye2, columns=['vo2max prediction'])
y_predict3.head()
```

```python
y_real = pd.DataFrame(y)
y_real = y_real.reset_index(drop=True)
y_real.head()
```

```python
dfsheet = pd.concat([y_real, y_predict3], axis=1)
dfsheet.head()
```

```python
from sklearn.metrics import (
    r2_score,
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)
```

```python
# using navis baye
r2 = r2_score(predicted_navis_baye2,y)
accuray = accuracy_score(predicted_navis_baye2,y)
f1 = f1_score(predicted_navis_baye2, y, average="weighted")
print(r2)
print(accuray)
print(f1)
```
